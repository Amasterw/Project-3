```python
college_data2=pd.read_csv("../Data/Colleges_and_Universities.csv")
college_data2.head()
```

```python
college_data2.columns
college_data2[college_data2['NAME']=='STANFORD UNIVERSITY']
```

```python
cols=['NAME','ADDRESS','CITY','STATE','ZIP','TELEPHONE','POPULATION','LATITUDE','LONGITUDE','WEBSITE']
college_data2_df=college_data2[cols].copy()
college_data2_df.head()
```

```python
college_data1_df.columns
```

```python
#college_data2.astype({'ZIP': str}).dtypes
college_data2_df['ZIP'] = college_data2_df['ZIP'].astype(str)
```

```python
college_data2_df.dtypes
```

```python
college_data=pd.merge(college_data1_df,college_data2_df,how='inner', on=['NAME','CITY'])
college_data.columns
```

```python
college_data.head(20)
```

```python
college_data.count()
```

```python
college_data.columns
```

```python
cols=['NAME','ADDRESS_x','CITY','STATE_y','ZIP_x','TYPE',
       'DEGREES OFFERED', 'CAMPUS SETTING', 'CAMPUS HOUSING',
       'STUDENT POPULATION', 'UNDERGRADUATE STUDENTS', 'PUBLIC/PRIVATE',
       'PUBLISHED IN-STATE TUITION AND FEES ($)',
       'PUBLISHED OUT-OF-STATE TUITION AND FEES ($)', 'TELEPHONE', 'POPULATION', 'LATITUDE', 'LONGITUDE', 'WEBSITE']
college_data_df=college_data[cols].copy()
```

```python
college_data_df.head()
college_data_df.count()
```

```python
college_data_df.rename(columns={"ADDRESS_x":"ADDRESS","STATE_y":"STATE","ZIP_x":"ZIP"}, inplace=True)
college_data_df.head()
```

```python
college_data_df[college_data_df["CITY"]=='BERKELEY']
```

```python
top_colleges=pd.read_csv("../Data/ranking.csv")
top_colleges.head()
```

```python
top_colleges=top_colleges.apply(lambda x: x.astype(str).str.upper())
top_colleges.columns= top_colleges.columns.str.upper()
top_colleges.head()
```

```python
college_data_final=pd.merge(college_data_df,top_colleges,how='left', on=['NAME'])
college_data_final.dtypes
college_data_final['RANK'] = college_data_final['RANK'].astype(float)
```

```python
college_data_final.dtypes
```

```python
college_data_final.sort_values(by=['RANK'], inplace=True)
```

```python
del college_data_final['STUDENT POPULATION_y']
```

```python
college_data_final.head()
```

```python
college_data_final.rename(columns={"STUDENT POPULATION_x":"STUDENT POPULATION"}, inplace=True)
college_data_final.head()
```

```python
college_data_final.to_csv("../Data/Final_College_Data.csv",index=False)
```
